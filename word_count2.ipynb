```
susan_words2 = []

for x in susan_words:
    for word in x:
        if word in STOPWORDS:
            continue
        else:
            susan_words2.append(word.lower())

len(susan_words2)
```

```
word_count_dict = {}

for word in susan_words2:
    if word in word_count_dict.keys():
        #increase value by 1
        word_count_dict[word] += 1
    else:
        #add word to keys
        word_count_dict[word] = 1

print(word_count_dict)
```
